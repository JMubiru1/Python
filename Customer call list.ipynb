```python
#df ["Phone_Number"].str.replace ("[^a-zA-Z0-9]", ".")
#df ["phone_Number"].apply (lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:10])
#df ["Phone_Number"]. apply (lambda x: str (x))
#df ["phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
#df

df ["Phone_Number"] = df ["Phone_Number"].str.replace ("N/a", " ")
df ["Phone_Number"] = df ["Phone_Number"].str.replace ("NaN", " ")
df ["Phone_Number"]. apply (lambda x: str (x))
df ["Phone_Number"] = df ["Phone_Number"].str.replace ("nan--", " ")

```

```python
df
```

```python
df[["Street_adress", "State", "Zip_code"]]=df["Address"].str.split(',' , expand = True)
df
```

```python
df = df.drop(columns=['Address'])
```

```python
df
```

```python
df['Paying Customer']=df['Paying Customer'].replace('Yes', 'Y')
df['Paying Customer']=df['Paying Customer'].replace('No', 'N')
df['Do_Not_Contact']=df['Do_Not_Contact'].replace('Yes', 'Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].replace('No', 'N')
df
```

```python
df=df.fillna(' ')
df
```

```python
df=df.replace ('N/a', '')
df
```
