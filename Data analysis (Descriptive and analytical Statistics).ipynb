DESCRIPTIVE STATISTICS
Objective: To determine the performance of different product lines (who buys what and where)

```python
df
```

1)Summary of the products

```python
product_counts = df['Product_Line'].value_counts()
product_percentages = df['Product_Line'].value_counts(normalize=True) * 100

product_summary = pd.DataFrame({
    'Count': product_counts,
    'Percentage (%)': product_percentages.round(2)
})

print(product_summary)
```

2) Products by country

```python
counts = df.groupby(['Country', 'Product_Line']).size().reset_index(name='Count')

total_per_country = counts['Count'].sum()

counts['Percentage (%)'] = (counts['Count'] / total_per_country * 100).round(2)

counts = counts.sort_values(by=['Country', 'Count'], ascending=[True, False])

print(counts)
```

3) Products by Gender

```python
counts = df.groupby(['Gender', 'Product_Line']).size().reset_index(name='Count')

total_count = counts['Count'].sum()

counts['Percentage'] = (counts['Count'] / total_count) * 100

counts = counts.sort_values(by=['Gender', 'Count'], ascending=[True, False])

print(counts)
```

```python
df['Gender'] = df['Gender'].replace({
    'M': 'male',
    'F': 'female',
    'Girl': 'female'
})
```

```python
gender_counts = df['Gender'].value_counts()

print(gender_counts)
```

```python
import matplotlib.pyplot as plt

counts = df.groupby(['Gender', 'Product_Line']).size().reset_index(name='Count')

total_count = counts['Count'].sum()

counts['Percentage'] = (counts['Count'] / total_count) * 100

counts['Label'] = counts['Gender'] 

plt.figure(figsize=(10, 10))
plt.pie(
    counts['Count'],
    labels=counts['Label'],
    autopct='%1.1f%%',
    startangle=140
)
plt.title('Distribution of Product Lines Grouped by Gender')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
