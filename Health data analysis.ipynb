```python
acceptable_ranges = {
    'Age': (0, 120),  
    'Billing Amount': (0, 1000000),  
    'Room Number': (0, 100000),  
}

for column, (min_value, max_value) in acceptable_ranges.items():
    if column in df.columns:
        out_of_range = df[(df[column] < min_value) | (df[column] > max_value)]
        if not out_of_range.empty:
            print(f"\n{column} has out-of-range entries:")
            print(out_of_range[[column]])
        else:
            print(f"\n{column} has no out-of-range entries.")
```

```python
max_age = df['Age'].max()
min_age = df['Age'].min()
mean_age = df['Age'].mean()
median_age = df['Age'].median()
q1 = df['Age'].quantile(0.25)  
q3 = df['Age'].quantile(0.75)  
iqr = q3 - q1

lower_whisker = max(min_age, q1 - 1.5 * iqr)
upper_whisker = min(max_age, q3 + 1.5 * iqr)

print("Max age:", max_age)
print("Min age:", min_age)
print("Mean age:", mean_age)
print("Median age:", median_age)
print("Lower quartile (Q1):", q1)
print("Upper quartile (Q3):", q3)
print("Interquartile Range (IQR):", iqr)
print("Lower whisker:", lower_whisker)
print("Upper whisker:", upper_whisker)
```

```python
age_stats = df.groupby('Medical Condition')['Age'].agg(
    mean='mean',
    std='std',
    median='median',
    iqr=lambda x: x.quantile(0.75) - x.quantile(0.25)  # Calculate IQR (75th percentile - 25th percentile)
)

print(age_stats)
```

6) relation between Billing Amount and Insurance Provider

```python
insurance_bills_sum = df.groupby('Insurance Provider')['Billing Amount'].sum().reset_index(name='total_bills')

total_bills = df['Billing Amount'].sum()

insurance_bills_sum['percentage_of_total'] = (insurance_bills_sum['total_bills'] / total_bills) * 100

insurance_bills_sum_sorted = insurance_bills_sum.sort_values(by='total_bills', ascending=False)

# 5. Print the sorted sum and percentage for each insurance provider
print("Sum and Percentage of Total Bills Per Insurance Provider (Descending Order):")
print(insurance_bills_sum_sorted)
```

```python
plt.figure(figsize=(8, 6))
plt.pie(insurance_bills_sum['total_bills'], 
        labels=insurance_bills_sum['Insurance Provider'], 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=['lightgreen', 'lightpink', 'lightcoral', 'lightblue', 'lightyellow'])

plt.title('Percentage of Total Bills Per Insurance Provider')

plt.axis('equal') 
plt.show()
```
